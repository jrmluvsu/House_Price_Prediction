```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
pd.pandas.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('train.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
features_with_na = [feature for feature in data.columns if data[feature].isnull().sum()>1]

for feature in features_with_na:
    print(feature, np.round(data[feature].isnull().mean(), 4), '% missing values')
```

```python
for feature in features_with_na:
    data1 = data.copy()
    data1[feature] = np.where(data1[feature].isnull(), 1, 0)
    
    data1.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
numerical_feature = [feature for feature in data.columns if data[feature].dtype != 'O']
```

```python
len(numerical_feature)
```

```python
numerical_feature
```

```python
data[numerical_feature].head()
```

```python
year_feature = [feature for feature in data.columns if 'Yr' in feature or 'Year' in feature]

year_feature
```

```python
for feature in year_feature:
    print(feature,':\n',data[feature].unique())
    print()
```

```python
data.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('YrSold')
plt.ylabel('SalePrice')
plt.title('House Price vs Year Of Sold')
```

```python
for feature in year_feature:
    data.groupby(feature)['SalePrice'].median().plot()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title('Different Year features with Sale Price')
    plt.show()
```

```python
for feature in year_feature:
    if feature != 'YrSold':
        data1 = data.copy()
        data1[feature] = data1['YrSold'] - data1[feature]
        plt.scatter(data1[feature], data1['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
discrete_feature = [feature for feature in numerical_feature if len(data[feature].unique())< 25 and feature not in year_feature + ['Id']]
discrete_feature
```

```python
len(discrete_feature)
```

```python
data[discrete_feature].head()
```

```python
for feature in discrete_feature:
    data1 = data.copy()
    data1.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.title('House Price vs Discrete features')
    plt.show()
```

```python
cont_feature = [feature for feature in numerical_feature if feature not in discrete_feature and feature not in year_feature + ['Id']]
cont_feature
```

```python
len(cont_feature)
```

```python
for feature in cont_feature:
    data1 = data.copy()
    data1[feature].hist(bins = 25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()
```

```python
for feature in cont_feature:
    data1 = data.copy()
    if 0 in data1[feature].unique():
        pass
    else:
        data1[feature] = np.log(data1[feature])
        data1['SalePrice'] = np.log(data1['SalePrice'])
        plt.scatter(data1[feature], data1['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

```python
for feature in cont_feature:
    data1 = data.copy()
    if 0 in data1[feature].unique():
        pass
    else:
        data1[feature] = np.log(data1[feature])
        data1.boxplot(column = feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
