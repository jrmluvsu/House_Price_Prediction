```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, data['SalePrice'], test_size = 0.1, random_state = 0)
```

```python
## for categorical features.
feature_na  = [feature for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtypes == 'O']
feature_na
```

```python
for feature in feature_na:
    print(feature, ':', np.round(data[feature].isnull().mean(), 4))
```

```python
def replace_cat_feature(data, feature_na):
    data1 = data.copy()
    data1[feature_na] = data1[feature_na].fillna('Missing')
    return data1
data = replace_cat_feature(data, feature_na)
data[feature_na].head()
```

```python
data[feature_na].isnull().sum()
```

```python
num_feature = [feature for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtypes != 'O']
num_feature
```

```python
for feature in num_feature:
    print(feature, ':', np.round(data[feature].isnull().mean(), 4))
```

```python
for feature in num_feature:
    median_value = data[feature].median()
    data[feature + 'nan'] = np.where(data[feature].isnull(), 1, 0)
    data[feature].fillna(median_value, inplace = True)

data[num_feature].isnull().sum()
```

```python
data.head(10)
```

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    data[feature] = data['YrSold'] - data[feature]
```

```python
data[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

```python
num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']
for feature in num_features:
    data[feature] = np.log(data[feature])
```

```python
data.head()
```

```python
cat_feature = [feature for feature in data.columns if data[feature].dtypes == 'O']
cat_feature
```

```python
for feature in cat_feature:
    ratio  = data.groupby(feature)['SalePrice'].count()/len(data)
    f_ind = ratio[ratio>0.01].index
    data[feature] = np.where(data[feature].isin(f_ind), data[feature], 'rare')
```

```python
data.head(20)
```

```python
for feature in cat_feature:
    labels_ordered=data.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    data[feature]=data[feature].map(labels_ordered)
```

```python
data.head()
```

```python
scaling_feature = [feature for feature in data.columns if feature not in ['Id', 'SalePrice']]
len(scaling_feature)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data1 = scaler.fit_transform(data[scaling_feature])
data1
```

```python
data = pd.concat([data[['Id', 'SalePrice']].reset_index(drop = True), pd.DataFrame(data1, columns = scaling_feature)], axis = 1)
```
