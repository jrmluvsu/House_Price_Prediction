```python
drop_feature = []
for feature in train_df.columns:
    if train_df[feature].isnull().sum() > 0.5*len(train_df):
        drop_feature.append(feature)
        train_df = train_df.drop(feature, axis = 1)
train_df.shape
```

```python
test_df = test_df.drop(['Alley', 'PoolQC', 'Fence', 'MiscFeature'], axis = 1)
```

```python
# total number of categorical features = 39
categorical_feature = [feature for feature in train_df.columns if train_df[feature].dtypes == 'O']
print(len(categorical_feature))
print(categorical_feature)
```

```python
# total number of numerical features = 38
numerical_feature = [feature for feature in train_df.columns if train_df[feature].dtypes != 'O']
print(len(numerical_feature))
print(numerical_feature)
```

```python
discrete_feature = [ feature for feature in train_df.columns if train_df[feature].dtypes != 'O' and len(train_df[feature].unique())<25]
print(len(discrete_feature))
print(discrete_feature)
```

```python
continuous_feature = [ feature for feature in train_df.columns if train_df[feature].dtypes != 'O' and len(train_df[feature].unique())>25]
print(len(continuous_feature))
print(continuous_feature)
```

## Categorical feature Missing Value Handling (Mode)

```python
train_df[categorical_feature].head(20)
```

```python
test_df[categorical_feature].head()
```

```python
for feature in categorical_feature:
    train_df[feature] = train_df[feature].fillna(train_df[feature].mode()[0])
    test_df[feature] = test_df[feature].fillna(test_df[feature].mode()[0])
```

```python
train_df[categorical_feature].head(20)
```

```python
test_df[categorical_feature].head()
```

```python
sns.heatmap(train_df.isnull(), yticklabels = False, cbar = False)
```

```python
sns.heatmap(test_df.isnull(), yticklabels = False, cbar = False)
```

## Numerical feature missing value handling (Mean)

```python
train_df[numerical_feature].head(10)
```

```python
numerical_feature_test = numerical_feature.copy()
numerical_feature_test.remove('SalePrice')
```

```python
test_df[numerical_feature_test].head(10)
```

```python
for feature in numerical_feature_test:
    train_df[feature] = train_df[feature].fillna(train_df[feature].mean())
    test_df[feature] = test_df[feature].fillna(test_df[feature].mean())
```

```python
train_df[numerical_feature].head(10)
```

```python
test_df[numerical_feature_test].head(10)
```

## Categorical Feature change to value

```python
data = pd.concat([train_df, test_df], axis = 0)
```

```python
data.shape
```

```python
i = 0
for feature in categorical_feature:
    df = pd.get_dummies(data[feature], drop_first = True)
    data = data.drop([feature], axis = 1)
    if i == 0:
        final_data = df.copy()
    else:
        final_data = pd.concat([final_data, df], axis = 1)
    i = i+1
data = pd.concat([data, final_data], axis = 1)
```

```python
data.shape
```

```python
data = data.loc[:, ~data.columns.duplicated()]
```

```python
data.shape
```

```python
data.head()
```

```python
train_df = data.iloc[:1460, :]
test_df = data.iloc[1460:, :]
y_train = train_df['SalePrice']
train_df = train_df.drop(['SalePrice', 'Id'], axis = 1)
test_df = test_df.drop(['SalePrice', 'Id'], axis = 1)
```

```python
n_estimator = [0.1, 1, 10, 100, 200, 400, 800, 1600]
max_depth = [2, 3, 5, 10, 15, 20]
learning_rate = [0.01, 0.05, 0.1, 0.2, 0.5]
min_child_weight = [1, 2, 3, 4, 5]
booster = ['gbtree', 'gblinear', 'dart']

param_distribution = {
   'n_estimator': n_estimator,
    'max_depth': max_depth,
    'learning_rate': learning_rate,
    'min_child_weight': min_child_weight,
    'booster': booster
}
```

```python
import xgboost
cl = xgboost.XGBRegressor()
```

```python
parameter.fit(train_df, y_train)
```
