```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

pd.pandas.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('X_train.csv')
```

```python
data.head()
```

```python
y_train = data['SalePrice']
X_train = data.drop(['Id', 'SalePrice'], axis = 1)
```

```python
print(X_train.shape, y_train.shape)
```

```python
feature_select = SelectFromModel(Lasso(alpha = 0.00001, random_state = 0))
feature_select.fit(X_train, y_train)
```

```python
feature_select.get_support()
```

```python
selected_feature = X_train.columns[feature_select.get_support()]
```

```python
selected_feature
```

```python
len(selected_feature)
```

```python
X_train = X_train[selected_feature]
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
predictor = LinearRegression()
predictor.fit(X_train, y_train)
y_pred = predictor.predict(X_train)
```

```python
y_train = y_train.values
```

```python
error = (y_train - y_pred)**2
```

```python
len(error)
```

```python
compare = pd.DataFrame([y_train, y_pred, error])
```

```python
compare = compare.T
```
